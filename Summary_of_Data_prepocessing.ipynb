### **Eg. replace NaN with 0:**

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'value':[700,np.nan,500,np.nan]})
print(df)
```

### method 1: drop the rows

```python
# Keep only the rows with at least 7 non-NA values.
df.dropna(thresh=7)
```

```python
# Define in which columns to look for missing values.
df.dropna(subset=['BloodPressure', 'SkinThickness','Insulin','Glucose','BMI'])
```

```python
# drop the zeros row
non_zero_col = ['BloodPressure', 'SkinThickness','Insulin','Glucose','BMI']
for i in non_zero_col:
  df[i]=df[i].replace(0,np.nan)
#df.head()
data = df.dropna(subset = ['BloodPressure', 'SkinThickness','Insulin','Glucose','BMI'])
data.head
#df.head()
```

### methods 2: replace the NaN with mean/median/average

```python
non_zero_col = ['BloodPressure', 'SkinThickness','Insulin','Glucose','BMI']
for i in non_zero_col:
  mean = df[i].mean(skipna=True)
  df[i] = df[i].replace(np.nan,mean)
df.head()
```
